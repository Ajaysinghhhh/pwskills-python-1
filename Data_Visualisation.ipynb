```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset
n_samples = 100
x1 = np.random.normal(loc=0, scale=1, size=n_samples)
x2 = x1 * 0.5 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # x2 is correlated with x1
x3 = x1 * -0.3 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # x3 is negatively correlated with x1
x4 = np.random.normal(loc=0, scale=1, size=n_samples)  # x4 is independent

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Compute the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
#np.random.seed(20)
#data = {
#    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#    'Day': np.random.choice(range(1, 31), 100),
#    'Sales': np.random.randint(1000, 5000, 100)
#}
#df = pd.DataFrame(data)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


# Set a random seed for reproducibility
np.random.seed(20)

# Generate the sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap='YlGnBu', cbar_kws={"label": "Total Sales"})
plt.title('Sales Variation Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.xticks(rotation=45)  # Rotate month labels for better readability
plt.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function.   z=sin(√x^2 +y^2)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)  # Create a meshgrid for 3D surface

# Calculate z data based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set titles and labels
ax.set_title('3D Surface Plot of z = sin(√(x² + y²))')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable notebook output if using Jupyter
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = np.sin(x)  # Calculate sine of x

# Create a new Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

# Show the plot
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the import numpy as np.


import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable notebook output if using Jupyter
output_notebook()

# Generate random data
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points) * 100  # x values from 0 to 100
y = np.random.rand(n_points) * 100  # y values from 0 to 100
sizes = np.random.randint(10, 100, n_points)  # Random sizes
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n_points)  # Random colors

# Create a DataFrame to hold the data
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a new Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X Values', y_axis_label='Y Values', width=800, height=400)

# Add a scatter renderer with variable sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', alpha=0.6)

# Customize the legend
p.legend.title = 'Colors'
p.legend.location = 'top_left'

# Show the plot
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable notebook output if using Jupyter
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           width=800, height=400)

# Add a bar renderer
p.vbar(x='fruits', top='counts', source=source, width=0.9, color='orange')

# Show the plot
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
#data_hist = np.random.randn(1000)
#hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable notebook output if using Jupyter
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           width=800, height=400)

# Add a quad renderer to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color='navy', line_color='white', alpha=0.7)

# Show the plot
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Enable notebook output if using Jupyter
output_notebook()

# Generate the dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=600, height=600)

# Define a color map
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data['value']), high=max(data['value']))

# Add squares for the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
        line_color=None, fill_color=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
