```python
#Q3 Explain the following terms:
# The difference in np.array, np.asarray and np.asanyarray.
# The difference between Deep copy and shallow copy.

#### 1. **The difference between `np.array`, `np.asarray`, and `np.asanyarray`:**

# **`np.array()`**: This function creates a new NumPy array from an existing array or other data structures (like lists). By default, it makes a copy of the input data, but you can change this behavior using the `copy` parameter.

# Example:
import numpy as np
lst = [1, 2, 3]
arr = np.array(lst)
print(arr)


# np.asarray(): It converts the input into an array, but if the input is already an array, it won't create a new copy (it will return the existing array). This makes it more efficient when you don't need to modify the original
# data.

# Example:
arr = np.asarray(lst)  # Won't make a copy if lst is already a NumPy array

# **`np.asanyarray()`**: Similar to `np.asarray()`, but it tries to convert the input into a base class (like `ndarray`), except if the input is already a subclass of `ndarray`. For example, if the input is a matrix
# (a subclass of `ndarray`), `np.asanyarray()` will preserve that type, whereas `np.asarray()` would force it to be a basic `ndarray`.

# Example:
m= np.mat([[1, 2], [3, 4]])
arr = np.asanyarray(m)


#### 2. **The difference between deep copy and shallow copy:**

# Shallow Copy: A shallow copy creates a new object but does not create copies of nested objects inside the original object. Instead, it only copies references to them. This means modifications to the nested objects in the copied array will also affect the original array.

# Example:
import copy
lst = [[1, 2], [3, 4]]
shallow_copy = copy.copy(lst)
shallow_copy[0][0] = 100
print(lst)

# Deep Copy: A deep copy creates a new object and recursively copies all objects inside the original one. This means that changes made to the copied object do not affect the original object.

# Example:
deep_copy = copy.deepcopy(lst)
deep_copy[0][0] = 100
print(lst)
```

```python
#Q5 Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array, perform the following operations:
# Extract all even integers from the array.
# Extract all odd integers from the array.

import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, (5, 6))

# Extract all even integers from the array
even_integers = array[array % 2 == 0]

# Extract all odd integers from the array
odd_integers = array[array % 2 != 0]

# Output the original array, even integers, and odd integers
print("Original Array:\n", array)
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

```python
#Q6 Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:
# Find the indices of the maximum values along each depth level (third axis).
# Perform element-wise multiplication of two arrays.

import numpy as np

# Create a 3D array with random integers between 1 and 10, shape (3, 3, 3)
array_3d = np.random.randint(1, 11, (3, 3, 3))

# Find the indices of the maximum values along each depth level (axis=2)
max_indices = np.argmax(array_3d, axis=2)

# Create another 3D array of the same shape for element-wise multiplication
array_3d_2 = np.random.randint(1, 11, (3, 3, 3))

# Perform element-wise multiplication between the two arrays
elementwise_multiplication = np.multiply(array_3d, array_3d_2)

# Output the results
print("Original 3D Array:\n", array_3d)
print("\nIndices of Maximum Values along depth (axis=2):\n", max_indices)
print("\nSecond 3D Array:\n", array_3d_2)
print("\nElement-wise Multiplication of Two Arrays:\n", elementwise_multiplication)
```

```python
#Q14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a time-series dataset
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=15, high=35, size=100)
humidity = np.random.uniform(low=40, high=80, size=100)

# Creating a DataFrame
data = {'Date': dates, 'Temperature': temperature, 'Humidity': humidity}
df = pd.DataFrame(data)

# Plotting Temperature and Humidity on the same plot with different y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (¬∞C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Creating a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Setting the title of the plot
plt.title('Temperature and Humidity Over Time')

plt.show()
```

```python
#Q15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
# tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.

# b) Overlay a line plot representing the normal distribution's probability density function (PDF).

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

# d) Set the title of the plot as 'Histogram with PDF Overlay

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay the normal distribution PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)  # mean=0, std=1
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
#Q17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as
# 'Quadrant-wise Scatter Plot

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant of each point
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to categorize points
quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create the scatter plot with Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue=quadrants, palette='bright', s=100)

# Add axis labels
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Add title
plt.title('Quadrant-wise Scatter Plot')

# Add legend
plt.legend(title='Quadrants')

# Show plot
plt.show()
```

```python
#Q19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title
# as 'Random Categorical Bar Chart'.
# Step 1: Import necessary libraries
# Step 1: Import necessary libraries

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256
import random

# Enable Bokeh output in Jupyter Notebook (skip this if running in a script)
output_notebook()

# Step 2: Generate random categorical data and values
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(10, 100) for _ in categories]  # Random values for each category

# Create a ColumnDataSource for the data
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Step 3: Create a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', toolbar_location=None, tools="")

# Step 4: Create the bar chart with color mapping based on values
p.vbar(x='categories', top='values', width=0.9, source=source,
       fill_color=factor_cmap('categories', palette=Viridis256, factors=categories), legend_field="categories")

# Step 5: Add hover tooltips to display the exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize the appearance of the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
p.xgrid.grid_line_color = None  # Remove the vertical grid lines
p.y_range.start = 0  # Start the y-axis at 0

# Display the plot
show(p)
```
